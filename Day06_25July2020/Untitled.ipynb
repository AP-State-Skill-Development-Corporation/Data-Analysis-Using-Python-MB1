```python
df = sn.load_dataset("iris")
sn.jointplot(x = "p_length",y = "p_width",data = iris,kind = 'reg')
plt.show()
```

```python
a = sn.jointplot("s_length","s_width",data = iris)
```

```python
a = sn.jointplot("s_width","p_length",data = iris,kind = 'kde')
```

```python
import matplotlib.pyplot as plt
g = (sn.jointplot("s_length","s_width",data = iris,color = "k").plot_joint(sn.kdeplot,n_levels = 10))
```
